```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
```

```python
var
```

```python
sns.set(style='whitegrid')
sns.boxplot(x=var.day,y=var.total_bill,order=['Fri','Sun','Thur','Sat'],
            showmeans=True,meanprops={'marker':"+",'markeredgecolor':"b"} ,linewidth=3,palette='plasma')
```
